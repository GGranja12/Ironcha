````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    print("Error: Quantity cannot be negative. Please enter a valid quantity.")
                else:
                    inventory[product] = quantity
                    valid_quantity = True
            except ValueError:
                print("Error: Invalid input. Please enter a numeric value.")
    return inventory
````

````python
def calculate_total_price(inventory):
    total_price = 0
    for product, quantity in inventory.items():
        while True:
            try:
                price = float(input(f"Enter the price for {product}: "))
                if price < 0:
                    print("Price cannot be negative. Please enter a valid price.")
                else:
                    total_price += price * quantity
                    break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
    return total_price
````

````python
def get_customer_orders(inventory):
    customer_orders = {}
    while True:
        order = input("Enter a product name to order (or 'done' to finish): ").strip()
        if order.lower() == 'done':
            break
        if order not in inventory:
            print("Error: Product not found. Please enter a valid product name.")
            continue
        
        while True:
            try:
                quantity = int(input(f"Enter the quantity for {order}: "))
                if quantity <= 0:
                    print("Quantity must be a positive integer.")
                elif quantity > inventory[order]:
                    print("Not enough stock available. Please enter a valid quantity.")
                else:
                    if order in customer_orders:
                        customer_orders[order] += quantity
                    else:
                        customer_orders[order] = quantity
                    break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
    return customer_orders
````

````python
products = {
    "t-shirt": 10,
    "mug": 5,
    "hat": 7,
    "book": 3,
    "keychain": 15
}

inventory = initialize_inventory(products)
print("Inventory initialized:", inventory)

orders = get_customer_orders(inventory)
print("Customer Orders:", orders)

total_price = calculate_total_price(inventory)
print(f"Total Price: ${total_price:.2f}")
````
