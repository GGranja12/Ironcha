```python
import time

# Define suspects and the murderer
suspects = ["Guilherme", "Charlotte", "Ruben", "Oeendrila"]
murderer = "Ruben"

# Dictionary of clues for the rooms
clues_in_rooms = {
    'kitchen': {
        "clues": {
            1: 'Beer cup',
            2: 'Broken vodka bottle',
            3: 'Running shoes',
            4: 'Homemade sandwiches',
            5: 'Chinese Chef Knife',
        },
        "required_clues": {5},
        "accessible_rooms": {1: 'garage', 2: 'living_room'}
    },
    'living_room': {
        "clues": {
            1: 'Rope',
            2: 'Notebook',
            3: 'Mountain shoes',
            4: 'Leather Jacket',
            5: 'Cutters'
        },
        "required_clues": {4},
        "accessible_rooms": {1: 'kitchen', 2: 'garage', 3: 'bedroom', 4: 'bathroom', 5: 'Accusatory room'}
    },
    'bedroom': {
        "clues": {
            1: 'Laptop',
            2: 'Crime book',
            3: 'Glass of wine',
            4: 'Curtain cord',
            5: 'Handbag'
        },
        "required_clues": {1},
        "accessible_rooms": {1: 'bathroom'}
    },
    'garage': {
        "clues": {
            1: 'Cleaning gloves',
            2: 'Tennis racket',
            3: 'Poker cards',
            4: 'Gun',
            5: 'Football ball',
        },
        "required_clues": {3},
        "accessible_rooms": {1: 'living_room'}
    },
    'bathroom': {
        "clues": {
            1: 'Belt',
            2: 'Opened snacks',
            3: 'Drug tablet',
            4: 'Glasses'
        },
        "required_clues": {4},
        "accessible_rooms": {1: 'living_room'}
    },
    'Accusatory room': {
        "clues": {},
        "required_clues": set(),
        "accessible_rooms": {1: 'living_room'}
    }
}

# Initialize game state
game_state = {
    "current_room": 'living_room',
    "collected_clues": set(),
}

def print_clues(room):
    """Print all clues available in the room."""
    clues = clues_in_rooms[room]["clues"]
    print(f"\nYou are in the {room}. You are looking around, you observe many clues:")
    for number, clue in clues.items():
        print(f"- {number}: {clue}")
        time.sleep(2)  

def explore_room(room):
    """Explore the current room and collect clues."""
    print(" ")
    print(f"\nYou are now exploring the {room}.")
    print_clues(room)
    clue_number = input("Enter the clue number you want to collect or type 'exit' to leave: ").strip()
    
    if clue_number.lower() == 'exit':
        return
    
    try:
        clue_number = int(clue_number)
        if clue_number in clues_in_rooms[room]["clues"]:
            if clue_number in clues_in_rooms[room]["required_clues"]:
                game_state["collected_clues"].add(clue_number)
                print(f"Clue {clue_number} collected!")
            else:
                time.sleep(2)  
                print(" ")
                print("Oh no! Unfortunately, either the DNA was deleted or none of the suspects used this object. This clue is not helpful for your investigation. You have to go back to the living room to talk to your police partner again.")
                game_state["current_room"] = 'living_room'
        else:
            print("Invalid clue number.")
    except ValueError:
        print("Please enter a valid number.")

def enter_house():
    """Ask the player if they are ready to enter the house."""
    ready = input("Are you ready to enter the house and start the investigation? (yes/no): ").strip().lower()
    if ready == 'yes':
        game_state["current_room"] = 'living_room'
    else:
        print("Come back when you are ready.")
        exit()

def play_game():
    """Main game loop."""
    enter_house()
    
    print("\nYou are now in the living room.")
    
    while True:
        current_room = game_state["current_room"]
        time.sleep(2)
        print(f"\nYou are in the {current_room}.")
        
        if current_room == 'Accusatory room':
            if clues_in_rooms[current_room]["required_clues"].issubset(game_state["collected_clues"]):
                print("You have all the required clues. You can now make an accusation.")
                suspect = input("Who do you want to accuse? (Enter suspect name): ").strip()
                
                if suspect in suspects:
                    if suspect == murderer:
                        time.sleep(2)
                        print(" ")
                        print("Congratulations! You have correctly accused the murderer!")
                        time.sleep(2)
                        print(" ")
                        print("Ruben, driven by jealousy and desperation to outshine Dr.")
                        time.sleep(2)
                        print(" ")
                        print("Norbert Wiener in the field of data science, used the Chinese chef’s knife from the kitchen to commit the murder by strangulation.")
                        time.sleep(2)
                        print(" ")
                        print("Afterward, he moved the body to the garage, hoping to hide his crime.") 
                        time.sleep(2)
                        print(" ")
                        print("Overcome with remorse, he returned to the scene, feigning innocence and pretending he hadn’t been at the party.") 
                        time.sleep(2)
                        print(" ")
                        print("He seized the opportunity created by Oeendrila’s absence to act quickly and cover his tracks, believing he could still escape suspicion.") 
                        time.sleep(2)
                        print(" ")
                        print("Thanks to you, he'll spend the rest of his life in prison! Thank you for your help!") 
                        
                        break
                    else:
                        print("Wrong accusation. Try again.")
                else:
                    print("Invalid suspect name.")
            else:
                print("You do not have all the required clues to accuse anyone. Explore other rooms to collect more clues.")
                game_state["current_room"] = 'living_room'
                continue
            
        time.sleep(2)
        action = input("Do you want to 'explore' the room or 'move' to another room? (Enter '1' to explore or '2' to move): ").strip().lower()
        
        if action == '1':
            explore_room(current_room)
            
        elif action == '2':
            # Move to another room
            accessible_rooms = clues_in_rooms[current_room]["accessible_rooms"]
            next_room = input(f"Which room would you like to go to? ({', '.join(accessible_rooms.values())}): ").strip().lower()
            if next_room in accessible_rooms.values():
                game_state["current_room"] = next_room
            else:
                print("Invalid room.")
        else:
            print("Invalid action. Please choose '1' or '2'.")

play_game()
```
